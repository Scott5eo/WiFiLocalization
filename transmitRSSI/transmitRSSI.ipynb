```python
import serial
import json
import numpy as np

# Open serial port
ser = serial.Serial('/dev/ttyUSB0', 115200)

# Read data from serial port
while True:
    data = ser.readline()
    data = data.decode("utf-8")

    if "Waiting for input" in data:
        print("connected")
        break

# Send data to serial port
file_path = "test_data/test0_0_0.json"
macs = []
rssi = []
with open(file_path, "r") as f:
    #read the json file
    data = json.load(f)
    for d in data:
        d = json.loads(d)
        macs.append(d["MAC"])
        rssi.append(d["RSSI"])

#argsort rssi
rssi = np.array(rssi)
idx = np.argsort(rssi)[::-1] #holds index in descending order
rssi = rssi[idx]

#get top 10
rssi = rssi[:5]

#normalize rssi values to range 0~20 and round to nearest integer
rssi = np.interp(rssi, (rssi.min(), rssi.max()), (0, 20))
rssi = np.round(rssi).astype(int)

cmd = ""
for i in range(len(rssi)):
    j = idx[i]
    #numpy int64 rssi number to string
    rss = str(rssi[i])
    cmd += macs[j] + ";" + rss + ","

#send cmd to serial port
ser.write(cmd.encode("utf-8"))


print("sent")
while True:
    data = ser.readline().decode("utf-8")
    print(data)
```
